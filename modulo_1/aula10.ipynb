# Aula 10 - Programação Orientada a Objetos

- Paradigma que visa a criação e manipulação de objetos
- Objetos são instancias de uma classe
- Uma classe é uma estrutura que define a o comportamento e as propriedades de um objeto
-  Uma classe define algo que já existe no mundo real
- Os comportamentos são escritos como métodos e funções



```python
class Pessoa:   # nome da classe sempre com letra maiúscula
 
    # O primeiro método sempre vai ser o que define quais são os atributos
 
    def __init__(self, nome, idade, altura):     # Função Construtora que é usada para definir os atributos da classe
        self.nome = nome
        self.idade = idade
        self.altura = altura
                                            # Os objetos também possuem ações, que são definidas pelos métodos/funções
    def apresentar(self):
            print(f'Olá, meu nome é {self.nome}, tenho {self.idade} anos e {self.altura} de altura')


p1 = Pessoa("Cássio",33,1.77)           #Defino os objetos
p2 = Pessoa("Amanda",31,1.65)

p1.apresentar()         #Chamo a ação
p2.apresentar()


```

## Proteção de acesso

- Encapsular significa agrupar e proteger métodos dentro de uma classe
- O encapsulamento permite a proteção dos dados a serem trabalhados, facilitando manutenção e evitando erros

```python
# Entendendo no código:

class Pessoa:   
 
    def __init__(self, nome, idade, altura):     
        self.__nome = nome                                  # para encapsular, basta acrescentar __ depois do ponto (.)
        self.idade = idade
        self.altura = altura
                                            
    def apresentar(self):
            print(f'Olá, meu nome é {self.__nome}, tenho {self.idade} anos e {self.altura} de altura')  #Altero também o caminho pro atributo sempre que ele é chamado.

    # Para acessar de novo o atributo que foi encapsulado, precisamos criar um método get():

    def get_nome(self):
         return self.__nome

p1 = Pessoa("Cássio",33,1.77)           
p2 = Pessoa("Amanda",31,1.65)

p1.apresentar()         
p2.apresentar()


# Pra chamar de novo o atributo nome, chamo pelo método get:

print(p1.get_nome())
```

## Proteção de modificação:

- O Encapsulamento também impede que alterações sejam feitas nas variáveis

```python
# Entendendo no código:

class Pessoa:   
 
    def __init__(self, nome, idade, altura):     
        self.__nome = nome                                  
        self.__idade = idade                              # definimos o __ antes do nome do atributo no método construtor
        self.altura = altura
                                            
    def apresentar(self):
            print(f'Olá, meu nome é {self.__nome}, tenho {self.idade} anos e {self.altura} de altura')  

    def get_nome(self):
         return self.__nome
    
    # Para alterar, criamos a função set():

    def set_idade(self, nova_idade)                         # Definimos uma variável para armazenar o novo atributo
         self.__idade = nova_idade                          # Declaramos para alterar o atributo

p1 = Pessoa("Cássio",33,1.77)           
p2 = Pessoa("Amanda",31,1.65)

p1.apresentar()         
p2.apresentar()


print(p1.get_nome())
```

## Herança

- É quando uma classe filha "herda"  os atributos de uma classe mãe.


```python
# Entendendo no código:

class Pessoa:   
 
    def __init__(self, nome, idade, altura):     
        self.__nome = nome                                  
        self.__idade = idade                              
        self.altura = altura
                                            
    def apresentar(self):
            print(f'Olá, o nome do pessoa é {self.__nome}, tem {self.__idade} anos e {self.altura} de altura')  

    def get_nome(self):
         return self.__nome

    def set_idade(self, nova_idade):                         
         self.__idade = nova_idade   

# Na herança, defino a classe e de qual ela é filha                       

class Aluno(Pessoa):                  
    def __init__(self, nome, idade, altura, matricula):                 # Defino o método construtor com novo atributo
        super().__init__(nome, idade, altura)               # super() permite acessar os atributos do construtor da classe mãe
        self.matricula = matricula                      # Defino o atributo novo

    # Função para imprimir matricula do aluno

    def imprimir_matricula(self):
     print(f'Matrícula do aluno: {self.matricula}.')

# Cadastro de aluno

aluno1 = Aluno('Cassio', 33, 1.77, '005258')               

aluno1.apresentar()             # método da classe mãe (Pessoa)
aluno1.imprimir_matricula()     # método da classe filha (Aluno)


```

# Polimorfismo

- Permite de diferentes classes tenham métodos com o mesmo nome
- Na prática, permite que as classes filhas consigam herdar também os métodos das mães

```python
# Entendendo no código:

class Pessoa:   
 
    def __init__(self, nome, idade, altura):     
        self.__nome = nome                                  
        self.__idade = idade                              
        self.altura = altura
                                            
    def apresentar(self):
            print(f'Olá, o nome do pessoa é {self.__nome}, tem {self.__idade} anos e {self.altura} de altura')  

    def get_nome(self):
         return self.__nome

    def set_idade(self, nova_idade):                         
         self.__idade = nova_idade   
                     

class Aluno(Pessoa):                  
    def __init__(self, nome, idade, altura, matricula):                 
        super().__init__(nome, idade, altura)               
        self.matricula = matricula                      

    def imprimir_matricula(self):
     print(f'Matrícula do aluno: {self.matricula}.')

    # Método que a filha vai herdar da classe mãe 

    def apresentar(self):
        print(f'O nome do aluno é {super().get_nome()}. Sua matrícula é {self.matricula}.')  # Uso o método super() para linkar com o nome, mas lembrando de acessar pelo get_nome, uma vez que encapsulamos o nome raiz da construtora.



aluno1 = Aluno('Cassio', 33, 1.77, '005258')               

aluno1.apresentar()            
aluno1.imprimir_matricula()     
```
