# Exercícios aula 5

- Crie uma lista de alunos e realize as seguintes operações:

    1. Adicione novos alunos que estão fazendo esta aula;
    2. um dos alunos desistiu da aula, agora você precisa remover ele da lista;
    3. Por fim, use o conceito de dicionário para que você possa pesquisar uma aluno em específico nessa lista


```python
# Criar a lista
alunos = [("Maria" , "15 anos"), ("Leticia" , "16 anos"), ("Lucas" , "15 anos"), ("Vanessa" , "14 anos"), ("Marcelo" , "15 anos")]
print(f'Lista de alunos da aula de matemática: {alunos}')

#Adicionar novos alunos

novo_aluno = ("Maicon" , "15 anos")
alunos.append(novo_aluno)
print(f'Lista de alunos atualizada com novo aluno: {alunos}')

#Remover um dos alunos

alunos.pop(1)
print(f'Lista atualizada com desistência: {alunos}')

#transformar em dict e pesquisar um aluno

alunos_dict = dict(alunos)

print(alunos_dict["Vanessa"])

```

# Exerciocios aula 7

1. Crie uma lista e use a declaração da lista;
2. Depois aplique os princípios de pilha, fila e lista que aprendemos nessa aula, adicionando e inserindo frutas;
3. utilize o laço de repetição para imprimir todos elementos dessa lista
4. Por fim, crie uma condição para imprimir somente a maça e laranja. Você só pode imprimir maça e laranja.   

```python
#Criar lista

frutas = []
frutas.append("Uva")
frutas.append("Limão")
frutas.append("Maçã")
frutas.append("Laranja")
frutas.append("Amora")

print(f'Lista completa: {frutas}')
print()

frutas.pop()
print('No modo convencional, em pilha, o último adicionado é deletado com o .pop(): ')
print(frutas)
print()

# transformar em fila

from collections import deque

frutas_fila = deque(frutas)
frutas_fila.popleft()
print('No modo fila, removemos o primeiro adicionado com o .popleft(): ')
print(frutas_fila)
print()

# imprimir os elementos da lista usando repetição
# criar condição e imprimir maça e laranja

print('Frutas da lista com laço for e condicional para "Maçã" e "Laranja": ')
for fruta in frutas_fila:
    if fruta == "Maçã":
        print(fruta)
    elif fruta == "Laranja":
        print(fruta)






```

3. em seguida, crie os seguintes métodos: mostrar_placa() e ao instanciar a classe a partir de um objeto, mostre a placa  deste veículo;

```python
class Carro:
    def __init__(self, modelo, placa, ano):
        self.modelo = modelo
        self.placa = placa
        self.ano = ano

    def mostrarPlaca(self):
        print(f'Placa do carro: {self.placa}')

car1 = Carro("Onix" , "ACB-2525" , 2023)
car2 = Carro("Uno" , "DFG-5859", 2005)

car1.mostrarPlaca()
car2.mostrarPlaca()
```
