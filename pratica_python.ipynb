## Exercícios aula 2

- Escreva um pequeno programa que peça ao usuário que digite seu nome e sua idade;
- O programa deve calcular e imprimir quantos anos ele terá em 2030;

```python
nome = input("Digite o seu nome: ")
idade = int(input("Em que ano você nasceu?: "))

idade2030 = (idade - 2030) * (-1)

print(f'Você se chama {nome} e em 2030 terá {idade2030} anos de idade!')
```
