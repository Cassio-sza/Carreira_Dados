```python
# Utilizando

import matplotlib.pyplot as plt

# Construindo gráfico de linhas

plt.plot([1,2,3], [4,9,3])

plt.show()

# Construindo um gráfico de barras
    # Dados
x = ['Uva', 'Amora', 'Abacaxi']
y = [8, 4, 5]

plt.bar(x, y)

    # Rótulos do gráfico
plt.xlabel('Frutas')
plt.ylabel('Quantidade')
plt.title('Quantidade de frutas')
```
