- Crie um dataframe que contenha algumas linhas e colunas: Nome, Endereço, Data de Nascimento, Data de admissão,Salário e Cargo;
- Em seguida, mostre na telatodas as linhas da coluna de data de admissão;

```python
# Criar dataframe

data = {
    'Nome': ['Alberto', 'Ana Paula', 'Milena', 'Leandro', 'Marcos'],
    'Endereço': ['Goiânia', 'Belo Horizonte', 'São Paulo', 'Salvador', 'Vitória'],
    'Data_de_nascimento': ['25/08/1974', '17/08/1991', '15/04/1995', '16/01/1984', '15/12/1989'],
    'Data_de_admissão': ['02/03/1999', '09/11/2002', '05/08/2019', '01/05/2015', '22/02/2024'],
    'Salário': [4450.00, 4900.00, 3875.00, 4100.00, 2950.00],
    'Cargo': ['Sênior', 'Sênior', 'Pleno', 'Pleno', 'Junior']
}

df = pd.DataFrame(data)

print(f'Dataframe:\n{df}\n')

# Mostrar a coluna da data de admissão

print(f'Coluna Data de admissão:\n{df['Data_de_admissão']}')
```

## Aula 2

- Crie um array com alguns valores;
- Em seguida, mostre na tela a média desses valores;

```python
#Importar biblioteca:
import numpy as np

#Criar array:
arr1 = pd.array([5, 15, 21, 35, 41, 52, 67, 68, 75, 88, 89, 95])
print(f'Array:\n{arr1}\n')

#Média do array:
media_array = np.mean(arr1)
print(f'Média do array:\n{media_array}')
```

1. Crie dois arrays com nomes e salários de funcionários;
2. Em seguida, mostre na tela o gráfico destes valores; 

```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Arrays

arr1 = np.array(['Matheus', 'Vitória', 'Giovana', 'Ana Paula'])
arr2 = np.array([4500, 3900, 5000, 6200])

plt.bar(arr1, arr2)

plt.xlabel('Funcionário')
plt.ylabel('Salário')

```
