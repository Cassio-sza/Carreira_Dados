# Desafio Gemini

1. Crie um array com o nome de 5 itens do estoque (ex: Luvas, Soro, Propofol, Gaze, Agulhas).

2. Crie um segundo array com as quantidades atuais desses itens.

3. Calcule a reposição: Imagine que o estoque ideal de cada item é 100 unidades. Crie um terceiro array que seja o resultado de: 100 - array_de_quantidades.

4. Visualize: Gere um gráfico de barras mostrando quanto de cada item precisa ser comprado (o resultado do terceiro array).

```python
import numpy as np
import matplotlib.pyplot as plt

item = np.array(['Ciprofloxacino', 'Dipirona', 'Sonda', 'Seringa', 'Furosemida'])
quantidade = np.array([23, 78, 96, 12, 33])
reposicao = 100 - quantidade

print(reposicao)

plt.bar(item, reposicao, color='black')
plt.title('Reposição CAF')

plt.xlabel('Material')
plt.ylabel('Quantidade a repor')

plt.show()




```
