```python
import pandas as pd
import numpy as np
```

```python
#Importar o dataset

df = pd.read_csv('/home/cassio/Documentos/programação/Senai/Carreira_Dados/modulo_2/praticas/archive/Ice Cream Sales - temperatures.csv')

#Informações primárias
df.head(10)
df.info()
```

```python
# Buscar nulos
print(df.isnull().sum())

# Buscar duplicados
duplicates = df[df.duplicated()]
print(len(duplicates))
```

```python
# Tratar duplicados
df.drop_duplicates(keep='last', inplace=True)
print(len(duplicates))
```

```python
# Renomear colunas
df = df.rename(columns={'Temperature': 'Temperatura', 'Ice Cream Profits': 'Lucro'})

df.info()
```

```python
# ordenar pela temperatura
sorted = df.sort_values(by='Temperatura', ascending=True)
print(sorted)
```
