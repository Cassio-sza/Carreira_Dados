# Git, GitHub e versionamento

- É a maneira mais fácil e aceita na comunidade para versionamento dos código;
- Open source e de viés contributivo, permite que além de guardar nosso códigos, propor mudanças e melhorias em códigos de outro programadores
- Permite a recuperação de etapas anteriores do código no caso de necessidade
- Permite que mais de uma pessoa mexa no código através de versões, que podem ou não serem incorporadas no código raiz (pull request)
- Sistema mais utilizado para armazenar e compartilhar código, uma vez que todas as versões e modificações são guardadas.
- Funciona através de clones do código
- Além disso, é o jeito mais fácil de armazenar os dados na nuvem e protege-los, uma vez que ele não depende só do seu computador 

- Git é o sistema e GitHub é a 'rede social' dos códigos

# 🛠️ Guia Rápido de Git para Desenvolvedores

Este resumo contém os comandos essenciais para o fluxo de trabalho em Ciência de Dados e IA.

---

## 1. Configuração Inicial
> Execute apenas uma vez na máquina ou ao trocar de perfil.

* `git config --global user.name "Seu Nome"`
* `git config --global user.email "seu@email.com"`

---

## 2. Iniciando e Clonando
* `git init`: Transforma a pasta atual em um repositório Git.
* `git clone <url>`: Copia um repositório remoto para sua máquina.

---

## 3. Fluxo de Trabalho Diário (O "Caminho Feliz")
1. **Verificar status:** `git status`
2. **Adicionar mudanças:** `git add .` (adiciona tudo) ou `git add arquivo.py`
3. **Salvar versão (Commit):** `git commit -m "Explicação breve da alteração"`
4. **Enviar para a Nuvem:** `git push origin main`

---

## 4. Ramos (Branches) e Colaboração
* `git branch`: Lista as branches existentes.
* `git checkout -b <nome-da-branch>`: Cria e entra em uma nova branch.
* `git checkout <nome>`: Alterna entre branches existentes.
* `git pull origin main`: Atualiza seu código local com o que está no servidor.
* `git merge <nome>`: Une o código da branch `<nome>` à branch onde você está.

---

## 5. Desfazendo Erros (Emergência)
* `git checkout -- <arquivo>`: Descarta mudanças não salvas em um arquivo específico.
* `git reset HEAD~1`: Desfaz o último commit, mas mantém os arquivos alterados.
* `git clean -df`: Remove arquivos novos que não foram rastreados (cuidado!).

---
