# Data Storytelling: Criando narrativas

- Do inglês, storytelling é contar história
- Data storutelling é a arte de contar histórias através da análise de dados

- Dados -> Informações -> insights
- Coletar -> Organizar

## Coletando dados

- Existem dois tipos diferentes de pesquisa de coleta de dados:
    - Quantitativa:É uma pesquisa em que vamos buscar dados numéricos acerca do fato 
        - **Ex: Com que frequência você assiste filmes?** - Essa resposta provavelmente virá em valor númerico, em dias

    - Qualitativa: Tipo de pesquisa que usamos a experiência do usuário para embasar os fatos. Menos estruturado e mais focado em um uúnico problema
        - **Ex: De todos os filmes que já assistiu, qual foi o que mais gostou?**

- A análise dos dados será focada basicamente em responder perguntas através dos dados de uma forma ou de outra, a depender de cada caso
        


### Como fazer boas perguntas?

- Existem 4 hacks para fazer boas perguntas:

    1. **Faça perguntas claras.** 
        - Escolha palavras simples e objetivas. Evite palavras dúbias e redundantes, evitando dúvida em em vai ver a análise
    
    2. **Mantenha um número equilibrado de opções de respostas.**
        - Crie um  padrão para a sua pesquisa. As repetiçãos de padrões geram familiaridade; Facilita na hora das respostas

    3. **Seja imparcial nas perguntas.**
        - o objetivo do questionário é fazer com que o outro seja sincero sobre o que ele acha, e não o que vocẽ acha;

    4. **Crie perguntas diferentes umas das outras**
        - Deixe o questionário mais atrativo alternando entre questões de múltipla escolha, abertas ou escala de níveis;

### Técnicas de storytelling

- Quem?
    - Quem é o seu público? O que ele faz? O que ele quer ver? - Saber pra quem você vai aplicar apara que vocẽ consiga apresentar da melhor forma

- O quê?
    - Mesmo que você já conheça o púplico, é importante saber o que ele consome, quais seus hábitos para saber o que apresentar para ele. Saber o que precisa ser apresentado para não trazer o qe ele não precisa ou não quer ver.

- Como?
    - Como seus dados vão ser apresentados pro seu público de modo que ele tenha a informação a ser passada da forma mais clara possível.

## Como organizar os dados

Representações visuais para apresentação dos dados:

- **Textos simples;**
- **Tabelas;**
- **Cores de forma moderada:** Técnica de mapa de calor;
- **Gráficos de pontos:** Para dados quantitativos;
- **Gráfico de linhas:** Representam evolução temporal;
- **Gráfico de barras:** Seervem para represenar comparações;
- **Gráfico de pizza:** Representa categorias distintas de uma pergunta;

- Independene da técnica de apresentação de dados que escolhermos usar, é importante ter em mente que os dados são o elemento mais importante da apresentação.
