## Mineração e preparação



- Mineração é o ato de buscar dados em banco de dados com a finalidade de tirar insights a partir de sua análise;
- As etapas da mineração são:
    - Definição do problema,
    - COleta de dados,
    - Pré processamento e preparação dos dados,
    - Análise exploratória,
    - Modelagem,
    - avaliação dos resultados.
    
- Pandas é uma das bibliotecas mais usadas para análise de dados;
- Numpy é a biblioteca que auxilia na parte matemática da análise, principalmente quando usamos números 
- Os dados podem conter erros, informações faltantes e desconexas. Por isso, é importante a etapa da coleta, limpeza e preparação dos dados.
- As etapas da preparação dos dados são:
    - Limpeza;
        Quando ocorre a remoção de dados duplicados, remoção de erros e o preenchimento de dados ausentes
    - Transformação;
        Acontece a conversão de formatos, normalização e padronmização dos dados
    - Integração
- para a coleta de dados, usaremos a biblioteca pandas
- No pandas as estruturas de dataframes e de series
- Os dataframes funcionam como uma espéecie de planilha
- As series se assemelham com arrais dimensionais e listas

## Pandas



- A estrutura mais importante que usaremos é a de dataframes
- São organizados m linhas e colunas
- Para usar um dataframe em python podemos usar o comando pd.dataframe()
- Podemos passar como argumento bibliotecas, listas ou arrays do numpy


```python
# Criando um dataframe 

data = {
    'Nome': ['Alice', 'Bruno', 'Matheus', 'David'],
    'Idade': [25, 34, 28, 31],
    'Cidade': ['Vitória', 'Florianópolis', 'Recife', 'Belo Horizonte']
}

df = pd.DataFrame(data)

print(f'Dataframe:\n{df}')
print('-' * 50,'\n')

# Para ver uma chave específica, chamo ela como argumento em df['nome_da_chave']:
print(f'Visualizando uma chave específica:\n{df['Nome']}\n')    

# Para ver uma linha específica, comando df.iloc[index_da_linha]:
print(f'Linha 1:\n{df.iloc[1]}\n') 

# Para ver um valor específico, defino o index e a chave com o comando df.loc[index, 'nome_da_chave']:
print(f'Index 1 e chave Cidade:\n{df.loc[1, 'Cidade']}')


```
